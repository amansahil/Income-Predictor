## Clustering

```python
# Spliting the data set into two subsets based on income
over_50k = dataset[dataset["Income"] == " >50K"]
under_50k = dataset[dataset["Income"] == " <=50K"]
```

### Age vs Hours Per Week

```python
X_over_50k = over_50k.iloc[:, [0, 12]].values
X_under_50k = under_50k.iloc[:, [0, 12]].values
```

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X_over_50k)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method for income over 50k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X_under_50k)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method for income under 50k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Fitting the data
kmeans_over_50k = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans_over_50k = kmeans_over_50k.fit_predict(X_over_50k)

kmeans_under_50k = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans_under_50k = kmeans_under_50k.fit_predict(X_under_50k)
```

```python
# Visualising the data
plt.scatter(X_over_50k[y_kmeans_over_50k == 0, 0], X_over_50k[y_kmeans_over_50k == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X_over_50k[y_kmeans_over_50k == 1, 0], X_over_50k[y_kmeans_over_50k == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X_over_50k[y_kmeans_over_50k == 2, 0], X_over_50k[y_kmeans_over_50k == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans_over_50k.cluster_centers_[:, 0], kmeans_over_50k.cluster_centers_[:, 1], s = 100, c = 'yellow', label = 'Centroids')
plt.title('Clusters for income over 50k')
plt.xlabel('Age')
plt.ylabel('Hours Per Week')
plt.legend()
plt.show()

plt.scatter(X_under_50k[y_kmeans_under_50k == 0, 0], X_under_50k[y_kmeans_under_50k == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X_under_50k[y_kmeans_under_50k == 1, 0], X_under_50k[y_kmeans_under_50k == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X_under_50k[y_kmeans_under_50k == 2, 0], X_under_50k[y_kmeans_under_50k == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans_under_50k.cluster_centers_[:, 0], kmeans_under_50k.cluster_centers_[:, 1], s = 100, c = 'yellow', label = 'Centroids')
plt.title('Clusters for income under 50k')
plt.xlabel('Age')
plt.ylabel('Hours Per Week')
plt.legend()
plt.show()
```

## Data Preprocessing

```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
# Enconding Categorical Data and convertion to array
X = pd.get_dummies(X, columns=[
    "WorkClass", "Education", "MaritalStatus", "Occupation", "Relationship",
    "Race", "Sex", "NativeCountry",
]).values

# Enconding with LabelEncoder due to white space
from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y.values)
```

```python
# Spliting Dataset into the Training set and Test set
from sklearn.model_selection import train_test_split # Deprecated use model_validation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Feature Scailing
from sklearn.preprocessing import MinMaxScaler
mm_X = MinMaxScaler()
X_train = mm_X.fit_transform(X_train)
X_test = mm_X.transform(X_test)
```

```python
# Applying LDA to reduce dimensions
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
pca = LDA(n_components = 2)
X_train_LDA = pca.fit_transform(X_train, y_train)
X_test_LDA = pca.transform(X_test)
```

## Ensemble Stacking

```python
from sklearn.model_selection import StratifiedKFold
def Stacking(model, train, y, test, n_fold):
    folds = StratifiedKFold(n_splits = n_fold)
    test_pred = np.empty((0,1), float)
    train_pred = np.empty((0,1), float)
    
    for train_indices, val_indices in folds.split(train, y.values):
        x_train, x_val = train.iloc[train_indices], train.iloc[val_indices]
        y_train, y_val = y.iloc[train_indices], y.iloc[val_indices]

        model.fit(X = x_train.values, y = y_train.values.ravel())
        train_pred = np.append(train_pred, model.predict(x_val))
        
    test_pred = np.append(test_pred, model.predict(test))    
    return test_pred.reshape(-1,1), train_pred
```

```python
X_test_stacking = pd.DataFrame(X_test_LDA)
X_train_stacking = pd.DataFrame(X_train_LDA)
y_train_stacking = pd.DataFrame(y_train)

from sklearn.svm import SVC
model1 = SVC(C = 100, gamma = 0.8, kernel = 'rbf')

test_pred1, train_pred1 = Stacking(model = model1, n_fold = 10, train = X_train_stacking, test = X_test_stacking, y = y_train_stacking)

train_pred1 = pd.DataFrame(train_pred1)
test_pred1 = pd.DataFrame(test_pred1)

from sklearn.naive_bayes import GaussianNB
model2 = GaussianNB()

test_pred2, train_pred2 = Stacking(model = model2, n_fold = 10, train = X_train_stacking, test = X_test_stacking, y = y_train_stacking)

train_pred2 = pd.DataFrame(train_pred2)
test_pred2 = pd.DataFrame(test_pred2)

from sklearn.tree import DecisionTreeClassifier
model3 = DecisionTreeClassifier(criterion = 'gini', max_depth = 8, min_samples_leaf = 5, min_samples_split = 4, min_weight_fraction_leaf = 0.0)

test_pred3, train_pred3 = Stacking(model = model3, n_fold = 10, train = X_train_stacking, test = X_test_stacking, y = y_train_stacking)

train_pred3 = pd.DataFrame(train_pred3)
test_pred3 = pd.DataFrame(test_pred3)
```

```python
df = pd.concat([train_pred1, train_pred2], axis=1)
df_test = pd.concat([test_pred1, test_pred2], axis=1)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# Applying Grid Search
from sklearn.model_selection import GridSearchCV
parameters = [
    {
        'C': np.logspace(-4, 4, 50),
        'penalty': ['l1', 'l2'],
        'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
    }
]

grid_search = GridSearchCV(estimator = model, param_grid = parameters, scoring = 'accuracy', cv = 3, n_jobs = -1)
```

```python
grid_search.fit(df, y_train_stacking.values.ravel())
y_pred = grid_search.predict(df_test)
```
